```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Initialize K-Means with the number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit K-Means to the data
kmeans.fit(X)

# Get cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.title('K-Means Clustering (Iris Dataset)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
cluster_counts = np.bincount(labels)
plt.pie(cluster_counts, labels=['Cluster 0', 'Cluster 1', 'Cluster 2'], autopct='%1.1f%%', startangle=140)
plt.title('Cluster Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(range(len(cluster_counts)), cluster_counts, color='skyblue')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.title('Cluster Counts')
plt.xticks(range(len(cluster_counts)), ['Cluster 0', 'Cluster 1', 'Cluster 2'])
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# True labels of the Iris dataset (for reference)
true_labels = iris.target

# Create a confusion matrix between true labels and cluster labels
conf_matrix = confusion_matrix(true_labels, labels)

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Cluster')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
